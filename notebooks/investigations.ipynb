```python
import pandas as pd  # Import the library
import socket

# Ensure the database engine is connected (if not already)
if 'engine' not in locals():
    from sqlalchemy import create_engine
    import os
    from dotenv import load_dotenv
    load_dotenv('../.env')
    # Reconnect just in case
    db_user = os.getenv("POSTGRES_USER", "hunter")
    db_pass = os.getenv("POSTGRES_PASSWORD", "password123")
    db_name = os.getenv("POSTGRES_DB", "monero_network")
    engine = create_engine(f"postgresql://{db_user}:{db_pass}@localhost:5432/{db_name}")

# Now fetch the data
query = "SELECT * FROM nodes;"
df = pd.read_sql(query, engine)

print(f"ðŸ“Š Loaded {len(df)} nodes.")

# Now run the investigation
suspicious_nodes = df[df['isp_name'].str.contains("DigitalOcean", na=False)]
print(f"ðŸ•µï¸ Investigating {len(suspicious_nodes)} suspicious nodes...\n")

for index, row in suspicious_nodes.head(20).iterrows():
    ip = row['ip']
    try:
        hostname = socket.gethostbyaddr(ip)[0]
        print(f"IP: {ip} -> Hostname: {hostname}")
    except:
        print(f"IP: {ip} -> No Hostname found")
```
